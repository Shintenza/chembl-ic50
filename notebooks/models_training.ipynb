```python
import sys
import os
import glob
import torch.nn as nn
import torch.optim as optim
```

```python
sys.path.append(os.path.abspath("../src"))
from utils.consts import DATA_FILE_PREFIX, BASELINE_DATA_LOCATION, SPLIT_MAP_NAME
from models.baseline_mlp import train_baseline_mlp
from utils.evaluation import evaluate_model
```

```python
batch_files = glob.glob(f"../{BASELINE_DATA_LOCATION}/{DATA_FILE_PREFIX}*.parquet")
splits_location = f"../{BASELINE_DATA_LOCATION}/{SPLIT_MAP_NAME}"
```

```python
model = nn.Sequential(
    nn.Linear(2048, 512),
    nn.ReLU(),
    nn.Dropout(p=0.3),
    nn.Linear(512, 128),
    nn.ReLU(),
    nn.Dropout(p=0.3),
    nn.Linear(128, 1)
)
```

```python
train_baseline_mlp(
    batch_files,
    splits_location,
    model,
    nn.MSELoss(),
    optim.Adam(model.parameters(), 0.001),
)
```

```python
evaluate_model(model)
```
